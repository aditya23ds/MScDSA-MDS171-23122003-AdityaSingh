```python


class ExpenseTracker:
# Initializing the list
    def __init__(self):
        self.transactions ={
            "EXPENSE":[],
            "INCOME":[]
        }

    def clear_transactions(self): 
        self.transactions ={
            "EXPENSE":[],
            "INCOME":[]
        }
        print("All transactions cleared")

# Load the CSV file and populate the dictionary
    def CsvFile(self):
        with open("D:\\github\\MScDSA-MDS171-23122003-AdityaSingh\\lab09.csv", "r") as csvfile:
            line=csvfile.readlines()
        for l in line[1:]:
            row = l.strip().split(',')
            if len(row)==5:
                type=row[0]
                category=row[1]
                cost=int(row[2])
                desc=row[3]
                date=row[4]
                self.StoreTransactions(type,category,cost,desc,date)
                # print("added the transcation")
            else:
                print("not added")

# Storing the transactions
    def StoreTransactions(self,type,category,cost,desc,date):
        transaction={
            "CATEGORY":category,
            "COST":cost,
            "DESC":desc,
            "DATE":date
        }
        if type=="EXPENSE":
            self.transactions['EXPENSE'].append(transaction)
        else :
            self.transactions['INCOME'].append(transaction)
        return True
    
    def view_transactions(self):
        print("INCOME")
        for abc in self.transactions['INCOME']:
            print(abc)
        print("EXPENSES")
        for abc in self.transactions['EXPENSE']:
            print(abc)

    def get_total_income(self):
        total_income=0
        for i in self.transactions['INCOME']:
            total_income+=i['COST']
        print("THE Total Income:",total_income)

    def get_total_expense(self):
        total_expense=0
        for i in self.transactions['EXPENSE']:
            total_expense+=i['COST']
        print("THE Total Expense:",total_expense)

    def exportToFile(self):
        with open("D:\\github\\MScDSA-MDS171-23122003-AdityaSingh\\lab09.csv", "a+") as file:
            for i in self.transactions['INCOME']:
                file.write(str(i)+'\n')
            for i in self.transactions['EXPENSE']:
                file.write(str(i)+'\n')
            print("exported to file successfully")

def get_input():
    category=input("Enter the category")
    cost=int(input("Enter the amount"))
    desc=input("Enter the description")
    date=input("Enter the date")
    return category,cost,desc,date

trac=ExpenseTracker()
trac.CsvFile()
trac.view_transactions()
trac.get_total_expense()
trac.get_total_income()

```

```python
trac.clear_transactions()

while True:
    print("*"*30)
    print("1.Add Income")
    print("2.Add Expense")
    print("3.Export")
    print("4.Exit")
    print("*"*30)

    i=int(input("Enter the choice"))

    if i==1:
        print("Enter the income details")
        category,cost,desc,date = get_input()
        trac.StoreTransactions('INCOME',category,cost,desc,date)
    elif i==2:
        print("Enter the expense details")
        category,cost,desc,date = get_input()
        trac.StoreTransactions('EXPENSE',category,cost,desc,date)
    elif i==3:
        print("Export")
        trac.exportToFile()
    elif i==4:
        print("THANK YOU !!!")
        break
    else:
        print("Invalid Choice")
```
